```
# Sorting Algorithm: Bubble Sort
def bubble_sort(arr):
  """
  Sorts a list using the Bubble Sort algorithm.

  Args:
    arr: The list to be sorted.

  Returns:
    The sorted list.
  """
  n = len(arr)
  # Traverse through all array elements
  for i in range(n):
    # Last i elements are already in place
    for j in range(0, n - i - 1):
      # traverse the array from 0 to n-i-1
      # Swap if the element found is greater than the next element
      if arr[j] > arr[j + 1]:
        arr[j], arr[j + 1] = arr[j + 1], arr[j]
  return arr
```

## Visualize with focus on
### Data ink maximization
Data-Ink Ratio = (Ink used to display data) / (Total ink used in the graphic)
#### Best Practices
- Remove chartjunk: Gridlines, borders, 3D effects, shadows
- Avoid redundant labels: Dont repeat axis titles or legends unnecessarily
- Simplify axes: Use minimal ticks and clean labels
- Use direct labeling: Label data points instead of relying on legends

### Graphical Design Principles
These principles guide the creation of truthful, efficient, and elegant visualizations:
1. Graphical Integrity
  - Avoid misleading scales (e.g., non-zero baselines)
  - Ensure proportional representation of values
2. Data Density
  - Maximize the amount of data shown without clutter
  - Use small multiples or layered plots for comparison
3. Avoid Chartjunk
  - Decorative elements that dont convey data (e.g., clipart, excessive colors)
  - Focus on data, not decoration
4. Contextual Clarity
  - Provide enough context to interpret the data
  - Use annotations, titles, and captions wisely
5. Multivariate Thinking
  - Show multiple variables when relevant (e.g., scatter plots with hue, size, shape)


### 1. Loading and Exploring the Data

We will be using a dataset of animal migration patterns. The dataset is in a CSV file named `bird_migration.csv`. It contains the following columns:

*   `bird_name`: The name of the animal species.
*   `latitude`: The latitude of the animal's location.
*   `longitude`: The longitude of the animal's location.

First, we need to import the necessary libraries and load the data into a pandas DataFrame. This is a crucial first step in any data analysis or visualization task, as it allows us to access and manipulate the data in a structured way.

```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import geopandas as gpd
from shapely.geometry import Point

df = pd.read_csv('bird_migration.csv')
df.head()
```

### 2. Visualizing Data with Seaborn

#### 2.1 Scatter Plots
A scatter plot is a fundamental tool for visualizing the relationship between two numerical variables. Each point on the plot represents an observation from the dataset. By plotting `longitude` vs. `latitude`, we can see the geographical distribution of the animal sightings. This gives us a good initial look at the locations, but it doesn't show us the path of migration. We can see different species are in different areas, which is a good start to our story.

```
sns.scatterplot(x='longitude', y='latitude', hue='bird_name', data=df)
plt.title('Animal Migration Scatter Plot')
plt.show()
```

#### 2.2 Line Plots
To better visualize the migratory patterns, a line plot is ideal. It connects the data points for each bird, showing the actual journey. This is a clear improvement over the scatter plot for understanding the migration paths. From now on, we'll use line plots to represent these paths as we explore how to customize our visualizations.

```
sns.lineplot(x='longitude', y='latitude', hue='bird_name', data=df)
plt.title('Animal Migration Line Plot')
plt.show()
```

### 3. Advanced Visualization: Mapping Animal Migrations
To give our data real-world context, we can plot the migration paths on a world map using `geopandas`. This creates a powerful and intuitive visualization of the animals' journeys, showing the story of their global travel.

```
import geodatasets
geometry = [Point(xy) for xy in zip(df['longitude'], df['latitude'])]
gdf = gpd.GeoDataFrame(df, geometry=geometry)

world = gpd.read_file(geodatasets.get_path('naturalearth.land'))

fig, ax = plt.subplots(figsize=(15, 10))
world.plot(ax=ax, color='lightgray')
gdf.plot(ax=ax, marker='o', color='red', markersize=5, label='Animal Location')
sns.lineplot(x='longitude', y='latitude', hue='bird_name', data=df, ax=ax)
plt.title('Animal Migration Paths')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.legend()
plt.show()
```

### 4. Distribution Plots
Now let's look at the distribution of our data. Where are these sightings concentrated?

#### 4.1 Histograms
A histogram visualizes the distribution of a single numerical variable. We can use it to see the distribution of animal sightings across different latitudes, helping us understand where the birds tend to congregate.

```
sns.histplot(x='latitude', data=df)
plt.title('Histogram of Latitude')
plt.show()
```

```
sns.histplot(x='longitude', data=df)
plt.title('Histogram of longitude')
plt.show()
```

#### 4.2 Kernel Density Estimates
A Kernel Density Estimate (KDE) plot is a smoothed version of a histogram. It's useful for visualizing the shape of a distribution. Here we can see the density of sightings by longitude.

```
sns.kdeplot(x='longitude', data=df)
plt.title('KDE Plot of Longitude')
plt.show()
```

```
sns.displot(data=df, x="longitude", y="latitude", hue='bird_name')
plt.title('Distribution Plot of Longitude and Latitude')
plt.show()
```

```
sns.pairplot(df, hue='bird_name')
plt.suptitle('Pairplot of Longitude and Latitude')
plt.show()
```

### 5. Categorical Plots

#### 5.1 Box Plots
A box plot is excellent for visualizing the distribution of a numerical variable across different categories. We can compare the latitudinal distribution for each bird species to see if their migration ranges differ.

```
sns.boxplot(x='bird_name', y='latitude', data=df)
plt.title('Box Plot of Latitude by Animal Name')
plt.show()
```

```
sns.boxplot(
    x='bird_name',
    y='latitude',
    hue='bird_name',
    data=df
)

plt.title('Box Plot of Latitude by Animal Name')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

#### 5.2 Violin Plots
A violin plot is a hybrid of a box plot and a KDE plot. It provides more information than a standard box plot by showing the probability density of the data. This is a powerful tool for comparing the distributions of longitude for each species.

```
sns.violinplot(
    x='bird_name',
    y='latitude',
    hue='bird_name',
    data=df
)

plt.title('Box Plot of Latitude by Animal Name')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```
# To generate a synthetic time series dataset and save it to a CSV file
# generate_time_series_data.py
import pandas as pd
import numpy as np

# Parameters
np.random.seed(42)
n_points = 1000
start_date = '2020-01-01'
end_date = '2022-01-01'

# Generate date range
date_range = pd.date_range(start=start_date, end=end_date, periods=n_points)

# Generate synthetic time series data
data = {
    'Date': date_range,
    'Value': np.sin(np.linspace(0, 50, n_points)) + np.random.normal(scale=0.5, size=n_points)
}

df = pd.DataFrame(data)
df.to_csv('time_series_data.csv', index=False)
print("Time series data generated and saved to 'time_series_data.csv'")
```

```
# To visualize the time series data using Matplotlib and Seaborn
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the time series data
df = pd.read_csv('time_series_data.csv', parse_dates=['Date'])

# Plot the time series data
plt.figure(figsize=(12, 6))
sns.lineplot(x='Date', y='Value', data=df, marker='o')
plt.title('Time Series Data')
plt.xlabel('Date')
plt.ylabel('Value')
plt.grid(True)
plt.savefig('time_series_plot.png')
plt.show()
```

```
#create an interactive visualization for exploring and querying time series data
import pandas as pd
import plotly.express as px

# Load the time series data
df = pd.read_csv('time_series_data.csv', parse_dates=['Date'])

# Create an interactive line plot
fig = px.line(df, x='Date', y='Value', title='Interactive Time Series Plot')
fig.update_xaxes(rangeslider_visible=True)
fig.show()
```

```
import pandas as pd
import plotly.express as px
from sklearn.manifold import TSNE
import numpy as np

# Example data creation
np.random.seed(0)
countries = ['USA', 'China', 'India', 'Brazil', 'Russia', 'Germany', 'UK', 'France', 'Italy', 'Canada']
years = range(2000, 2021)
data = []

for country in countries:
    for year in years:
        data.append({
            'Country': country,
            'Year': year,
            'Life Expectancy': np.random.uniform(60, 85),
            'GDP': np.random.uniform(10000, 50000),
            'Healthcare Expenditure': np.random.uniform(1000, 10000),
            'Funding': np.random.uniform(0, 1000)
        })

df = pd.DataFrame(data)
```

```
# Interactive Scatter Plot using t-SNE
features = df[['Life Expectancy', 'GDP', 'Healthcare Expenditure', 'Funding']].values
tsne = TSNE(n_components=2, random_state=0)
reduced_features = tsne.fit_transform(features)

scatter_df = pd.DataFrame(reduced_features, columns=['TSNE1', 'TSNE2'])
scatter_df['Country'] = df['Country']

fig_scatter = px.scatter(
    scatter_df,
    x='TSNE1',
    y='TSNE2',
    color='Country',
    title='t-SNE Visualization of Health Metrics'
)

fig_scatter.show()
```

```
# Time Series Line Chart
time_series_df = df.groupby(['Year']).agg({
    'Funding': 'sum'
}).reset_index()

fig_time_series = px.line(
    time_series_df,
    x='Year',
    y='Funding',
    title='Funding Patterns Over Time'
)

fig_time_series.show()
```

```
# create_and_populate_db.py
import pandas as pd
import numpy as np
import sqlite3
from sqlalchemy import create_engine
from sklearn.datasets import load_breast_cancer # Import the dataset
import os # Import os module to handle file operations

# Define database file path
db_file = 'multidimensional_data.db'

# Remove the database file if it exists to ensure a clean start
if os.path.exists(db_file):
    os.remove(db_file)
    print(f"Removed existing database file: {db_file}")

# Load a realistic high-dimensional dataset (30 features)
breast_cancer = load_breast_cancer(as_frame=True)
df = breast_cancer.frame
# Add target names for better interpretation in visualizations if needed
df['target_names'] = breast_cancer.target_names[breast_cancer.target]


# Connect to SQLite database
engine = create_engine(f'sqlite:///{db_file}', echo=False)

# Save DataFrame to database
df.to_sql('data', con=engine, if_exists='replace', index=False)

print("Database created and data populated with breast cancer dataset.")
```

```
# tsne_visualization.py
import pandas as pd
import numpy as np
from sklearn.manifold import TSNE
import seaborn as sns
import matplotlib.pyplot as plt
from sqlalchemy import create_engine

# Connect to SQLite database
engine = create_engine('sqlite:///multidimensional_data.db')

# Query the data
df = pd.read_sql("SELECT * FROM data", con=engine)
df.head()
print("Columns in DataFrame for t-SNE (static):", df.columns.tolist()) # Added for debug

# Apply t-SNE for dimensionality reduction, excluding the 'target' and 'target_names' columns from features
# 'target' is the numerical label (0 or 1), 'target_names' is the string label.
# We will use 'target_names' for coloring the plot.
tsne = TSNE(n_components=2, random_state=45)
tsne_results = tsne.fit_transform(df.drop(['target', 'target_names'], axis=1))

# Create a DataFrame for the t-SNE results and add the target_names for coloring
df_tsne = pd.DataFrame(tsne_results, columns=['Dimension_1', 'Dimension_2'])
df_tsne['target_names'] = df['target_names']

# Plot t-SNE results, colored by target class
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Dimension_1', y='Dimension_2', hue='target_names', data=df_tsne, alpha=0.7)
plt.title('t-SNE Visualization of Breast Cancer Data')
plt.savefig('tsne_plot.png')
plt.show()
```

```
# interactive_tsne.py
import pandas as pd
import numpy as np
from sklearn.manifold import TSNE
import plotly.express as px
from sqlalchemy import create_engine

# Connect to SQLite database
engine = create_engine('sqlite:///multidimensional_data.db')

# Query the data
df = pd.read_sql("SELECT * FROM data", con=engine)

print("Columns in DataFrame for t-SNE (interactive):", df.columns.tolist()) # Added for debug

# Apply t-SNE for dimensionality reduction, excluding the 'target' and 'target_names' columns from features
tsne = TSNE(n_components=2, random_state=42)
tsne_results = tsne.fit_transform(df.drop(['target', 'target_names'], axis=1))

# Create a DataFrame for the t-SNE results and add the target_names for coloring
df_tsne = pd.DataFrame(tsne_results, columns=['Dimension_1', 'Dimension_2'])
df_tsne['target_names'] = df['target_names']

# Plot interactive t-SNE results, colored by target class
fig = px.scatter(df_tsne, x='Dimension_1', y='Dimension_2', color='target_names', title='Interactive t-SNE Plot of Breast Cancer Data')
fig.show()
```

## Static 3D t-SNE Visualization

### Subtask:
Generate a static 3D t-SNE plot using Matplotlib. The t-SNE algorithm will reduce the high-dimensional data to 3 components, and the plot will show these three dimensions colored by the target class (malignant/benign).


**Reasoning**:
The subtask requires generating a static 3D t-SNE plot. I will implement the steps to connect to the database, perform t-SNE dimensionality reduction to 3 components, and then plot the results using matplotlib's 3D capabilities, coloring points by their target class.



```
import pandas as pd
import numpy as np
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sqlalchemy import create_engine

# 1. Connect to SQLite database
engine = create_engine('sqlite:///multidimensional_data.db')

# 2. Query the data
df = pd.read_sql("SELECT * FROM data", con=engine)

# 3. Instantiate TSNE with 3 components
tsne = TSNE(n_components=3, random_state=45)

# 4. Apply t-SNE for dimensionality reduction, excluding target and target_names
tsne_results = tsne.fit_transform(df.drop(['target', 'target_names'], axis=1))

# 5. Create a DataFrame for the t-SNE results
df_tsne_3d = pd.DataFrame(tsne_results, columns=['Dimension_1', 'Dimension_2', 'Dimension_3'])

# 6. Add the 'target_names' column for coloring
df_tsne_3d['target_names'] = df['target_names']

# 7. Create a 3D scatter plot
fig = plt.figure(figsize=(12, 10))
ax = fig.add_subplot(111, projection='3d')

# 8. and 9. Iterate through unique target names and plot each series
for target_name in df_tsne_3d['target_names'].unique():
    subset = df_tsne_3d[df_tsne_3d['target_names'] == target_name]
    ax.scatter(
        subset['Dimension_1'],
        subset['Dimension_2'],
        subset['Dimension_3'],
        label=target_name,
        alpha=0.7
    )

# 10. Set the title of the plot
plt.title('Static 3D t-SNE Visualization of Breast Cancer Data')

# 11. Add labels for the x, y, and z axes
ax.set_xlabel('Dimension 1')
ax.set_ylabel('Dimension 2')
ax.set_zlabel('Dimension 3')

# 12. Add a legend to the plot
ax.legend()

# 13. Display the plot
plt.show()
```

## Interactive 3D t-SNE Visualization

### Subtask:
Generate an interactive 3D t-SNE plot using Plotly Express. The t-SNE algorithm will reduce the high-dimensional data to 3 components, and the plot will show these three dimensions colored by the target class (malignant/benign), allowing for interactive exploration.

**Reasoning**:
The next step is to generate an interactive 3D t-SNE plot using Plotly Express, as outlined in the previous markdown cell. This involves connecting to the database, applying t-SNE to reduce dimensions to 3, and then using Plotly Express to create an interactive 3D scatter plot.



```
import pandas as pd
import numpy as np
from sklearn.manifold import TSNE
import plotly.express as px
from sqlalchemy import create_engine

# 1. Connect to SQLite database
engine = create_engine('sqlite:///multidimensional_data.db')

# 2. Query the data
df = pd.read_sql("SELECT * FROM data", con=engine)

# 3. Apply t-SNE for dimensionality reduction, excluding target and target_names
tsne = TSNE(n_components=3, random_state=42) # Using random_state=42 as in interactive_tsne.py
tsne_results = tsne.fit_transform(df.drop(['target', 'target_names'], axis=1))

# 4. Create a DataFrame for the t-SNE results
df_tsne_3d_interactive = pd.DataFrame(tsne_results, columns=['Dimension_1', 'Dimension_2', 'Dimension_3'])

# 5. Add the 'target_names' column for coloring
df_tsne_3d_interactive['target_names'] = df['target_names']

# 6. Plot interactive 3D t-SNE results, colored by target class
fig = px.scatter_3d(df_tsne_3d_interactive, x='Dimension_1', y='Dimension_2', z='Dimension_3',
                    color='target_names', title='Interactive 3D t-SNE Plot of Breast Cancer Data')
fig.show()
```
